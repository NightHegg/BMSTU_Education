## Задание 3
Напишите возмущенную систему линейных уравнений с учетом возмущения
входных данных. Как можно оценить относительную ошибку полученного
численного решения по сравнению с точным решением?

$(A + \delta A)(x + \delta x) = b + \delta b$ - возмущённая система уравнений.
\begin{equation}
\frac{\left \| \delta x \right \|}{\left \| \hat{x} \right \|} \leqslant \left \| A^{-1} \right \| \left \| A \right \| (\frac{\left \| \delta A \right \|}{\left \| A \right \|} + \frac{\left \| \delta b \right \|}{\left \| A \right \| \left \| \hat{x} \right \|}),
\end{equation}
где $\left \| A^{-1} \right \| \left \| A \right \| = \varkappa (A)$ - относительное число обусловленности.

***
**(Исправления!)**

Оценка погрешности возмущенного решения через невязку исходного уравнения:
$\frac{\left \| \delta x \right \|}{\left \| \hat{x} \right \|} \leqslant \frac{\left \| A^{-1} \right \| \cdot \left \| r \right \|}{\left \| \hat{x} \right \|}$, где $r = A \cdot \hat{x} - b$ - невязка.

## Задание 5
Что такое спектральный радиус матрицы R? Приведите условие сходимости
итерационного процесса в терминах спектрального радиуса. Приведите критерии
сходимости методов Якоби, Зейделя и последовательной верхней релаксации для
различных классов матриц.

Спектральным радиусом матрицы $R$ называется число
\begin{equation}
\rho(R) = max_i \left | \lambda_i \right |,
\end{equation}
где $\lambda_i$ - собственные значения матрицы $R$. 

Условие сходимости итерационного процесса в терминах спектрального радиуса:

Итерационный процесс $x_{m+1} = Rx_m + c$ сходится к решению системы $Ax=b$, каков бы ни был начальный вектор $\Leftrightarrow$ $\rho(R)<1$. 

Критерии сходимости:
- метод Якоби: $\rho(R_j)<1$, где $R_j = L + U$;
- метод Зейделя: $\rho(R_{GS})<1$, где $R_{GS} = (I-L)^{-1} U$; 
- последовательная верхняя релаксация: $\rho(R_{SOR})<1$, где $R_{SOR} = (I-\omega L)^{-1} ((1-\omega)I + \omega U)$, ($\omega >1$);

***
**(Исправления!)**

Критерии сходимости для разных матриц:
- если $A$ - матрица с строгим диагональным преобладанием по строкам, то методы Якоби и Гаусса-Зейделя сходятся, причём $\left \| R_{GS} \right \|_{\infty} \leqslant \left \| R_{J}  \right \|_{\infty} \leqslant 1$
- если $A$ - неразложимая матрица с слабым диагональным преобладанием по строкам, то методы Якоби и Гаусса-Зейделя сходятся, причём $\rho(R_{GS}) < \rho(R_{J}) < 1$
- если $A$ - симметричная положительно определенная матрица, то  $\rho(R_{SOR}) < 1$ для всех $\omega \in (0,2)$
- если $A$ - согласованно упорядоченная матрица, то  $\rho(R_{GS}) = \rho(R_{J})^{2}$

- если $A$ - согласованно упорядоченная матрица и $\omega \neq 0$, тогда:
    - $\mu$ - собственное значение матрицы $R_J$ и $(\lambda + \omega - 1)^2 = \lambda \omega^2 \mu^2$, где $\lambda$ - собственное значение матрицы $R_{SOR(\omega)}$;
    - если матрица $R_J$ имеет вещественные собственные значения и $\mu = \rho(R_J) < 1$, то

\begin{equation}
\rho(R_{SOR(\omega)}) = 
\left\{\begin{matrix}

\end{equation}
    где $\omega_{opt} = \frac{2}{1+\sqrt{1 - \mu^2}}$

## Задание 7
Что такое подпространство Крылова для матрицы A и вектора b? Что такое
алгоритм Ланцоша и какую информацию он позволяет получить на k -м шаге? Из
каких соображений можно строить алгоритмы с помощью крыловского
подпространства? Какая характеристика матрицы А влияет на сходимость
итерационного процесса в методе сопряженных градиентов? Как можно ускорить
процесс сходимости?

1) Подпространство Крылова $K_k(A, b)$ - линейная оболочка векторов $b, Ab, A^2 b$, ... $A^{k-1} b$.

2) Алгоритм Ланцоша - алгоритм для "частичного" приведения матрицы к симметричной трехдиагональной форме. После k шагов алгоритма в матрице T(которая, если А - симметричная матрица, то T = H, где H-верхняя хессенбергова матрица)
\begin{equation}
T = 
\begin{pmatrix}
T_k & T^T_{ku} \\
T_{ku} & T_u
\end{pmatrix}
\end{equation}
мы знаем блоки $T_k$ и $T_{ku}$, но не знаем блок $T_u$.

3) Алгоритмы можно строить с помощью крыловского подпространства с расчётом на то, что мы используем информацию, полученную на $k$ шагах алгоритмов Арнольди или Ланцоша. Мы также учитываем, что $k$ может быть меньше, чем $n$, что позволит провести эффективные расчёты.

4) Скорость сходимости итерационного процесса в методе сопряженных градиентов зависит от числа обусловленности матрицы A. Для ускорения процесса сходимости можно заменить систему $Ax=b$ системой $M^{-1}Ax=M^{-1}b$, где матрица $M$ есть приближение к $A$ со следующими свойствами:
- M симметрична и положительна определена;
- матрица $M^{-1}A$ хорошо обусловлена;
- система $Mx=b$ решается легко;

Данный выбор $M$ может сделать число обусловленности матрицы $M^{-1}A$ меньше, чем число обусловленности матрицы $A$, что ускорит сходимость.
