# Лекция 01
## Введение

### Основные методы, связанные с ОДУ

#### Определение 1. 
Обыкновенным дифференциальным уравнением (ОДУ) называется соотношение
\begin{equation}\label{eq:ref}
F(t, y(t), y'(t), \ldots, y^{(n)}(t)) = 0,
\end{equation}

связывающее значения независимой переменной $t$, искомой функции $y = y(t)$ и её производных до порядка $n \geq 1$. Порядок $n$ старшей производной, входящей в (\ref{eq:ref}), называется порядком уравнения.

#### Определение 2. 
Решение уравнения (\ref{eq:ref}) называется функция $y = y(t)$, определённая на некотором интервале (или отрезке), имеющая производные до порядка $n$ и удовлетворяющая уравнению (\ref{eq:ref}), т.е. для $\forall \: t$ из указанного интервала выполнено (\ref{eq:ref}).

#### Определение 3.
Если уравнение (\ref{eq:ref}), функция $F$ и $y$ зависят более чем от одной независимой переменной и в уравнение входят частные производные по этим переменным, то уравнение называется уравнением в частных производных (ДУЧП).

#### Определение 4. 
Уравнение (\ref{eq:ref}) называется разрешимым относительно старшей производной, если его можно представить в виде:
\begin{equation}\label{eq:ref_2}
y^{(n)} = f(t, y, y', \ldots, y^{(n-1)}
\end{equation}

В дальнейшем мы в основном будем рассматривать ОДУ 1-го порядка
\begin{equation}\label{eq:ref_3}
y' = f(t, y)
\end{equation}
и ОДУ 2-го порядка, т.е.
\begin{equation*}\tag{$3'$}
y''=f(t, y, y')
\end{equation*}

#### Определение 5.
Системой нормального вида называется система ОДУ
\begin{equation}\label{eq:ref_4}
\frac{dx_i}{dt} = f_i (t, x_1, \ldots, x_n), \ i = \overline{1, n}
\end{equation}
Число уравнение равно числу неизвестных функций $x_i(t)$.
Уравнение (\ref{eq:ref_3}) - частный случай (\ref{eq:ref_4}) для $n$=1.

Уравнение (\ref{eq:ref_2}) можно свести к (\ref{eq:ref_4}) заменой
\begin{equation*}
\begin{array}{ll}
x_1 = & y(t) \\
x_2 = & y'(t) = x_{1}'(t) \\
x_3 = & y''(t) = x_{2}'(t) \\
\vdots&  \\
x_n = & y^{(n-1)}(t) = x_{n-1}'(t)
\end{array}
\end{equation*}

Тогда получим систему
\begin{equation}
\left\{\begin{array}{ll}
\dfrac{dx_1}{dt} = & x_2 \\ 
\dfrac{dx_2}{dt} = & x_3 \\
\vdots \\
\dfrac{dx_n}{dt} = & f(t, x_1, x_2, \cdots, x_n)
\end{array}\right.
\end{equation}

В векторных обозначениях: $\dfrac{d\vec{x}}{dt} = \vec{f}(t, \vec{x})$, что совпадает с (\ref{eq:ref_3}) $\vec{x}=(x_1, \cdots, x_n)^T, \vec{f} = (f_1, \cdots, f_n)^T$.

Впервые векторные обозначения для системы ОДУ использовал Пеано в 1890 году.

Ещё Ньютон, Лейбниц и Эйлер заметили, что общее решение (\ref{eq:ref_3}) содержит 1 свободный параметр. Для уравнения (\ref{eq:ref_2}) и системы (\ref{eq:ref_4}) таких параметров $n$ штук. Чтобы выделить единственное решение, используются дополнительные соотношения.

Наиболее ранние известные случаи рассмотрения ОДУ
1. Ньютон 1671: $y' = 1-3x+y+x^2 + xy$
2. Лейбниц 1676: обратная задача о касательной (найти кривую $y(x)$ по заданной длине касательной)
\begin{equation*}
y' = -\dfrac{y}{\sqrt{a^2 - y^2}}
\end{equation*}
3. Иоганн Бернулли: задача о брахистохроне (наискорейший спуск) - начало вариационного исчисления (1696 г.)
\begin{equation*}
\int_{x_0}^{x_1} f(x, y, y')dx \rightarrow min
\end{equation*}
4. Эйлер 1744: $f_{y''y'}y'' + f_{y'y}y'+f_{y'x}-f_y=0$
5. Клеро 1734: неявные ОДУ $y - xy' +f(y') = 0$


#### Определение 6. 
Задача поиска решения системы уравнений
\begin{equation}\label{eq:ref_6}
\left\{\begin{array}{l}
y' = f(t, y) \\ 
y(0) = y_0
\end{array}\right.
\end{equation}
называется задачей Коши (ЗК) для уравнения 1 порядка. Аналогично для уравнения (\ref{eq:ref_2}) задача Коши имеет вид:
\begin{equation}\label{eq:ref_7}
\left\{\begin{array}{l}
y^{(n)} = f(t, y, y', \cdots, y^{(n-1)}) \\ 
y(0) = y_0 \\
y'(0) = y_1 \\
\vdots \\
y^{(n-1)}(0)= y_{n-1}
\end{array}\right.
\end{equation}

Для (\ref{eq:ref_4}) в векторной форме ЗК имеет вид:

\begin{equation}
\left\{\begin{array}{l}
\dfrac{d\vec{x}}{dt} = \vec{f}(t, \vec{x}) \\ 
\vec{x}(0) = \vec{x_0}
\end{array}\right.
\end{equation}

В случае, если поставлена задача вида

\begin{equation*}
\begin{array}{l}
y^{(n)} = f(t, y, y', \cdots, y^{(n-1)}) \\
y(a) = \varphi, y(b) = \psi, y'(a) = y_1, \cdots
\end{array}
\end{equation*}
то говорят о краевой задаче для ОДУ (это усложняет поиск решения).

Краевая задача для уравнения 2 порядка:
\begin{equation*}
\left\{\begin{array}{l}
y'' = f(t, y, y') \\ 
y(a) = \varphi, y(b) = \psi
\end{array}\right.
\end{equation*}

Рассмотрим ЗК (\ref{eq:ref_6}). Эйлер в 1768 году предложил следующий способ её решения:
- рассмотрим отрезок $\left [ 0, T \right ]$ и введём на нём разбиение $0 = t_0, \ t_1, \ldots, t_{n-1}, \ t_n=T$
- на каждом отрезке заменим решение $y(t_i)$ первым членом его ряда Тейлора:
\begin{equation*}
\begin{array}{l}
y(t_1) = y_1 = y_0 + (t_1 - t_0)y'(t_0)=y_0 + (t_1-t_0)f(t_0, y_0) \\
y(t_2) = y_2 = y_1 + (t_2 - t_1)y'(t_1)=y_1+(t_2-t_1)f(t_1, y_1) \\
\vdots \\
y(t_n)=y_n=y_{n-1} + (t_n - t_{n-1})y'(t_{n-1})=y_{n-1}+(t_n-t_{n-1})f(t_{n-1}, y_{n-1})
\end{array}
\end{equation*}

- если теперь соединить точки $(t_0, y_0), (t_1, y_1), \ldots, (t_n, y_n)$ прямыми, то получится ломаная Эйлера
\begin{equation*}
y_h(t) = y_i + (t-t_i)f(t_i, y_i) \ \mathrm{при} \ t \in [t_i, t_{i+1}] 
\end{equation*}
(по сути это и есть явный метод Эйлера для решения ОДУ)

Огромный вклад в теорию ОДУ внёс Коши. Около 1820 г. Коши сформулировал теорему сходимости для метода ломаных Эйлера, а позже ряд общих теорем существования решений дифференциальных уравнений. 

#### Определение 7.
Функция $\vec{f}(t, \vec{x})$ удовлетворяет условию Липшица по $\vec{x}$ на множестве $D$, если существует такая постоянная $\kappa$, что для каждых двух точек $(t, \vec{x}) \in D$ и $(t, \vec{y}) \in D$:
\begin{equation*}
\left \| \vec{f}(t, \vec{x}) - f(t, \vec{y}) \right \| \leq \kappa \left \| \vec{x} - \vec{y} \right \|
\end{equation*}

#### Лемма.
Если функция $f(t, \vec{x})$ имеет в выпуклой по $\vec{x}$ области частные производные $\dfrac{\partial f_i}{\partial x_j}$ и $\left | \dfrac{\partial f_i}{\partial x_j} \right | \leq L \: (i, j = \overline{1, n})$ в $D$, то $f(t,x)$ удовлетворяет условию Липшица в $D$ с $\kappa = n \cdot L$.

#### Теорема 1: 
Пусть на множестве $D = \left \{ (t, x): t_0 \leq t \leq T, \left | y-y_0 \right | \leq b \right \} \: f(t, y)$ - непрерывная функция, $\left | f \right | \leq A$, $f$ удовлетворяет условию Липшица. Если $T-t_0 \leq \frac{b}{A}$, то
1. При $\left | h \right | = max \left | t_{i+1} - t_i \right | \rightarrow 0$ ломаные Эйлера $y_n(t)$ сходятся равномерно к непрерывной функции $\varphi(t)$
2. $\varphi(t)$ - непрерывные дифференцируема и является решением ЗК (\ref{eq:ref_6}) на $\left [ t_0, T \right ]$
3. при $t \in \left [ t_0, T \right ]$ ЗК (\ref{eq:ref_6}) не имеет других решений

#### Теорема 2:
Пусть в области $D$ $\vec{f}(t, \vec{x})$ и её производные $\dfrac{\partial f_i}{\partial x_j} \ (i, j = \overline{1, n})$ непрерывны. Тогда для любой точки $(t_0, x_0) \in D$ может существовать не более одного решения ЗК
\begin{equation}\label{eq:ref_8}
\left\{\begin{array}{l}
\dfrac{d \vec{x}}{dt} = f(t, \vec{x}) \\
\vec{x}(t_0) = \vec{x_0}
\end{array}\right.
\end{equation}

На отрезке $I = \left [ t_0 - d, t_0 + d \right ]$, где $d = \frac{r}{\sqrt{m^2 + 1}} > 0$ ($r$ такое, что шар $s = \left \{ (t, \vec{x}) : (t-t_0)^2 + \left \| \vec{x} - \vec{x_0} \right \|^2 \leq r^2 \right \} < D$, $m=max\left \| \vec{f} \right \|$) решение задачи (\ref{eq:ref_8}) существует и единственно.

Рассмотрим краевую задачу:
\begin{align}
x_{1}'=f(t, x_1, x_2) \label{eq:ref_9_1} \tag{9.1} \\
x_{2}'=g(t, x_1, x_2) \label{eq:ref_9_2} \tag{9.2} \\
\varphi(x_1(a), x_2(a))=0 \label{eq:ref_9_3} \tag{9.3} \\
\psi(x_1(b), x_2(b))=0 \label{eq:ref_9_4} \tag{9.4}
\end{align}

Её решение можно свести к решению ЗК при помощи метода стрельбы:

1) Пусть $x_1(a) = \eta$ ($\eta$ - параметр)

2) Из (\ref{eq:ref_9_3}) находим $x_2(a)= \xi (\eta)$

3) Решим ЗК (\ref{eq:ref_9_1}), (\ref{eq:ref_9_2}) с начальными условиями $x_1(a)=\eta$, $x_2(a) = \xi(\eta)$ аналитически или численно. Получим $\tilde{x_1}(t, \eta)$, $\tilde{x_2}(t, \eta)$

4) На правом конце отрезка вычислим значения $\psi(\tilde{x_1}(b, \eta), \tilde{x_2}(b, \eta))$.

Решим уравнение $\psi(\tilde{x_1}(b, \eta), \tilde{x_2}(b, \eta))=0$ аналитическим или численно (с помощью решения нескольких задач Коши) с заданной точностью. Если правые части (\ref{eq:ref_9_1}) и (\ref{eq:ref_9_2}) удовлетворяют **Теореме 2**, а функция $\psi(\tilde{x_1}(b, \eta), \tilde{x_2}(b, \eta))$ имеет корни, то решение существует.

Рассмотрим ДУЧП вида $\dfrac{\partial \vec{u}}{\partial t} = \vec{f} (t, \vec{u}, \dfrac{\partial \vec{u}}{\partial x}, \dfrac{\partial^2 \vec{u}}{\partial x^2})$.
Часто решение этой системы может быть получено только численно. В этом случае удобно дискретизировать сначала правую часть (по переменной x) и свести ДУЧП к системе ОДУ.

#### Пример 
$\dfrac{\partial u}{\partial t} = a^2 \dfrac{\partial^2 u}{\partial x^2}$, $x \in \left [ 0, 1 \right ]$, $t \in \left [ 0, T \right ] $. 

\begin{array}{ll}
u = u(t, x) & u(0, x) = u_0(x) \\
u(t, 0) = \varphi_t & u(t, 1) = \psi(t)
\end{array}

Введём сетку $\omega_h = \left \{ 0 = x_0 < x_1 < x_2 < \ldots < x_n = 1 \right \}, x_i = i \cdot h$. Тогда запишем аппроксимацию 
\begin{equation}\label{eq:ref_10}
\frac{du_i}{dt} = a^2 \frac{u_{i+1} - 2u_i + u_{i-1}}{h^2}, \ i = 1, n-1
\end{equation}

Здесь для каждого узла сетки $i$ необходимо найти функцию $u_i(t)$. Если систему (\ref{eq:ref_10}) дополнить аппроксимацией граничных условий, то получим 
\begin{equation}\label{eq:ref_11}
\left\{
\begin{array}{l}
\dfrac{du_i}{dt} = a^2 \dfrac{u_{i+1} - 2u_i + u_{i-1}}{h^2}, \  i =\overline{1, n-1} \\
u_i(0)=u_0(x_i)
\end{array}
\right .
\end{equation}
с условиями $u_0(t)=\varphi(t)$, $u_n(t)=\psi(t)$. Задачу (\ref{eq:ref_11}) можно решить любым методом, предназначенным для ОДУ. Этот способ перехода от ДУЧП к ОДУ называется методом прямых или методом частичной дискретизации.

Ещё один тип задач для ОДУ - задача на СЗ или задача Штурма - Лиувилля. Впервые её рассмотрел Штурм в 1836 г. - задача определения свойств решения из самого уравнения (без получения непосредственно самого решения). 

В "Теории теплоты" Пуассона рассмотрено уравнение $\dfrac{\partial u}{\partial t} = \dfrac{\partial}{\partial x} (\kappa(x)\dfrac{\partial u}{\partial x}) - l(x) \ u$, $\kappa(x) > 0$. Пуассон использовал метод разделения переменных $u(x, t) = y(x) e^{-\lambda t}$, откуда $\dfrac{d}{dx} (\kappa(x) \dfrac{dy}{dx}) - l(x) \ y = -\lambda \ y'$ или
\begin{equation}\label{eq:ref_12}
(\kappa(x)y')' + G(x)y = 0
\end{equation}
где $G(x) = \lambda - l(x)$.

#### Теорема 3. 
Рассмотрим вместе с (\ref{eq:ref_12}) дифференциальное уравнение: 
\begin{equation}\label{eq:ref_13}
(\hat{\kappa}(x)\hat{y}')' + \hat{G}(x)\hat{y} = 0
\end{equation}

Пусть $\kappa, \hat{\kappa}$ дифференцируемы, $G$ и $\hat{G}$ непрерывны и для $\forall x$ выполнены соотношения $0 < \hat{\kappa}(x) \leq \kappa(x)$, $\hat{G}(x) \leq G(x)$. Пусть $y(x)$ и $\hat{y}(x)$ - линейно независимые (ЛНЗ) решения (\ref{eq:ref_12}) и (\ref{eq:ref_13}) соответственно. Тогда между любыми двумя нулями $y(x)$ находится по крайней мере один нуль $\hat{y}(x)$, т.е если $y(x_1) = y(x_2) = 0 (x_1 < x_2)$, то $\exists x_3 \in (x_1, x_2): \hat{y}(x_3) = 0$. 

Добавим к (\ref{eq:ref_12}) граничные условия
\begin{equation}\label{eq:ref_14}
y(x_0) = y(x_1) = 0
\end{equation}

#### Теорема 4.
Задача на СЗ (\ref{eq:ref_12}), (\ref{eq:ref_14}) имеет бесконенечную последовательность СЗ $\lambda_1 < \lambda_2 < \lambda_3 < \ldots$, которым отвечают решения $y_i(x)$ (собственные функции) с $0, 1, 2, \ldots$ нулями соответственно внутри интервала $(x_0, x_1)$. При этом нули функции $y_{j}(x)$ разделяются нулями функции $y_{j+1}(x)$. Если $0 < \kappa_1 \leq \kappa(x) \leq \kappa_2$ и $L_1 \leq l(x) \leq L_2$, то справедливы оценки $L_1 + \kappa_1 \dfrac{j^2 \pi^2}{(x_1 - x_0)^2} \leq \lambda_j \leq L_2 + \kappa_2 \dfrac{j^2 \pi^2}{(x_1 - x_0)^2}$.

### Устойчивость

#### Определение 8.
Пусть имеется система уравнений:
\begin{equation}\label{eq:ref_15}
y_i' = f_i(y_1, \ldots, y_n), i = \overline{1, n},
\end{equation}
    в которой $f_i(0, 0, \ldots, 0)=0, i=\overline{1, n}$ (поскольку $f_i$ не зависит в явном виде от $t$, но только от $y_1(t), \ldots, y_n(t)$, то система (\ref{eq:ref_15}) называется автономной). Тогда точка O называется устойчивой по Ляпунову, если $\forall \varepsilon >0: \exists \delta > 0$: для всех решений (\ref{eq:ref_15}) $\left \| y(x_0) \right \| < \delta \rightarrow \left \| y(x) \right \| < \varepsilon \ \forall x > x_0$. 
    
#### Определение 9.
Линеаризированная система для (\ref{eq:ref_15})
\begin{equation}\label{eq:ref_16}
y_i' = \sum_j a_{ij}, \ a_{ij} = \dfrac{\partial f_i}{\partial y_j}(0), i = \overline{1, n} 
\end{equation}

#### Теорема 5.
Система (\ref{eq:ref_16}) устойчива по Ляпунову $\Leftrightarrow$ все корни характеристического уравнения $det(\lambda I - A)=a_0\lambda + a_1 \lambda^{n-1} + \ldots +a_{n-1}\lambda + a_n=0$ удовлетворяют условию $Re \lambda=0$, а для кратных корней, порождающих при приведении $A$ к каноническому виду жордановы клетки, выполнено строгое неравенство $Re \lambda >0$ (это требование обусловлено тем, что решение (\ref{eq:ref_16}) ищется в виде разложения по функциям $e^{\lambda i t}$ (для простых корней) и $p_{k-1}(t)e^{\lambda i t}$ (для корня кратности k $p_{k-1}(t)$ - многочлен степени $k-1$).


## Глава 1. Методы Рунге-Кутты

Будем рассмматривать методы решения ЗК:
\begin{equation}\label{eq:ref_1_1}
\left\{\begin{array}{l}
u' = f(t, u) \\
u(0) = u_0 \\
t \in \left [ 0, T \right ]
\end{array}
\right .
\end{equation}
Эквивалентная ей задача 
\begin{equation}\label{eq:ref_1_2}
\int_{0}^{t} u(t) dt = \int_{0}^{t} f(t, u) dt \rightarrow u(t)=u_0 + \int_{0}^{t} f(t, u) dt
\end{equation}

Численные методы решения ЗК основаны на различных вариантах аппроксимации задач (\ref{eq:ref_1_1}) и (\ref{eq:ref_1_2}) - аппроксимациях производной $u'$, правой части $f(t, u)$, интеграла в (\ref{eq:ref_1_2}). Методы делятся на одношаговые (методы Рунге-Кутты), которые используют только одно предыдущее значение искомой функции для построения последующего и многошаговые. Кроме того, методы можно разделить на явные и неявные в зависимости от того, используется или нет при аппроксимации $f(t, u)$ значение функции $u$ в последующий момент времени.

Пусть $t_0 = 0$. Введём сетку $\omega_{\tau} = \left \{ t_k = k \cdot \tau, k = 0, 1, \ldots \right \} \ y_n=y(t_n)$ - сеточная функция.

#### Определение 1.1
Говорят, что численный метод обладает сходимостью в точке $t=t_n=n \cdot \tau$, если $\left | y_n -u(t_n) \right | \rightarrow 0$ при $\tau \rightarrow 0$ ($n \rightarrow \infty$). $y_n$ - приближенное решение, $u_n = u(t_n)$ - точное решение, $\left \{ y_n \right \}$-последовательность приближенных решений на последовательности сеток $\left \{ \omega_{\tau} \right \}$, $t_n=n \cdot \tau$.

#### Определение 1.2
Метод сходится на ($0, T$), если он сходится в любой точке $t \in (0, T)$.

#### Определение 1.3
Метод имеет $p$-ый порядок точности, если $\left | y_n - u(t_n) \right |=0 (\tau^{p})$ при $\tau \rightarrow 0$. 

#### Определение 1.4
Погрешностью метода называется сеточная функция $z_n = y_n - u(t_n)$.

### Методы Эйлера

Простейшие примеры одношаговых методов - методы Эйлера.

Явный метод. 
\begin{equation}\label{eq:ref_1_3}
\dfrac{y_{n+1} - y_n}{\tau} = f(t_n, y_n), n=0, 1, 2, \ldots \\
y_{n+1}=y_n + \tau f(t_n, y_n) \\
\dfrac{z_{n+1} - z_n}{\tau}=f(t_n, z_n+u(t_n)) - \dfrac{u_{n+1} - u_n}{\tau}=\psi_{h}^{(1)} + \psi_{h}^{(2)},
\end{equation}
где $\psi_{h}^{(1)} = f(t_n, u_n) - \dfrac{u_{n+1} -u_n}{\tau}$, $\psi_{h}^{(2)} = f(t_n, z_n + u_n) - f(t_n, u_n)$.

#### Определение 1.5
Функция $\psi_{h}^{(1)}$ - невязка или погрешность аппроксимации разностного метода на решении исходного ОДУ. 

#### Опредеделение 1.6
Говорят, что имеет место аппроксимация разностного уравнения на решении исходного ОДУ, если $\left | \psi_{h}^{(1)} \right | \rightarrow 0$ при $\tau \rightarrow 0$. 

Если $\psi_{h}^{(1)}=0 (\tau^p)$, то говорят об аппроксимации p-го порядка для (\ref{eq:ref_1_3}): $\psi_{h}^{(1)}=f(t_n, u_n) - u'(t_n) - \dfrac{1}{2} u''(t_n + O \tau) \tau = O(\tau)$.

#### Пример
Рассмотрим колебания груза на пружине (без трения): $m \ddot{x}=-kx$.

Система консервативна, энергия сохраняется $\rightarrow$
\begin{equation}\label{eq:ref_20}
\dfrac{m(\dot{x})^2}{2} + \dfrac{kx^2}{2}=\textrm{const} 
\end{equation}
Возьмём от выражения (\ref{eq:ref_20}) производную по времени:
\begin{equation}
m\dot{x}\ddot{x}+k\dot{x}x=0 \rightarrow \left [
\begin{array}{l}
\dot{x}=0 \\
m\ddot{x}+kx=0
\end{array}\right .
\end{equation}

Произведём замену $\dfrac{k}{m}=\omega^2 \rightarrow \ddot{x}+\omega^2=0$.

Сведём это уравнение к системе $u_1 = x, u_2 = \dot{u_1} = \dot{x}$:
\begin{equation}
\left\{\begin{array}{l}
\dot{u_1}=u_2 \\
\dot{u_2}=-\omega^2 u_1
\end{array}\right.
\end{equation}

Применим для решения явный метод Эйлера:
\begin{equation}
\left \{
\begin{array}{l}
y_{1, n+1}=y_{1,n}+\tau y_{2,n} \\
y_{2, n+1}=y_{2,n}-\tau \omega^2 y_{1,n}
\end{array}\right.
\end{equation}

Можно показать, что $\omega^2 y^2_{1, n+1} + y^2_{1, n+1}=(1+\tau^2 \omega^2)(y^2_{1,n} \omega^2 + y^2_{1, n}) > y^2_{1,n}\omega^2 + y^2_{2,n} \rightarrow$ энергия в системе возрастает!
